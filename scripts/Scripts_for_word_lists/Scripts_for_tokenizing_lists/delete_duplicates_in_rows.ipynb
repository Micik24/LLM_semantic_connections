```python
import csv

def remove_duplicates_from_csv(input_file, output_file):
    with open(input_file, newline='', encoding='utf-8') as infile, \
         open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        for row_number, row in enumerate(reader, start=1):
            seen = set()
            new_row = []
            for item in row:
                if item not in seen:
                    new_row.append(item)
                    seen.add(item)
            
            removed_count = len(row) - len(new_row)
            print(f"Row {row_number}: removed {removed_count} duplicates")

            writer.writerow(new_row)
```

```python

remove_duplicates_from_csv(r"NGSL_lists\Single_token_All_NGSL.csv", "output2.csv")
```
