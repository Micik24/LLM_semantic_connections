```python
import numpy as np
import matplotlib.pyplot as plt

# ---- 1. Load matrix ----
M = np.load(r"C:\Users\HP\bert_experiments\matrcies\Probabilities_NGSL.npy")   # change path if needed
print("Matrix shape:", M.shape)
print("dtype:", M.dtype)

# ---- 2. Basic statistics ----
print("Min:", np.min(M))
print("Max:", np.max(M))
print("Mean:", np.mean(M))
print("Std:", np.std(M))

# ---- 3. Histogram of all values ----
plt.figure(figsize=(8, 5))
plt.hist(M.ravel(), bins=100)   # ravel() flattens the matrix
plt.title("Value Distribution (Histogram)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# ---- 4. Heatmap (downsampled to avoid RAM issues) ----
# For 5000×5000, full heatmap can be heavy. Downscale to e.g. 500×500:
factor = 10  # 5000 / 10 = 500 pixels
M_small = M[::factor, ::factor]

plt.figure(figsize=(6, 6))
plt.imshow(M_small, aspect='auto')
plt.colorbar(label="Value")
plt.title(f"Heatmap (downsampled {factor}x)")
plt.show()
```
